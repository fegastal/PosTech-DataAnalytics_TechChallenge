```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ExpVinho = pd.read_csv('/content/ExpVinho.csv', sep=';', header=0)
df_ExpVinho
```

```python
colunas_nomes_anos = df_ExpVinho.columns[2:]
colunas_nomes_anos
```

```python
dict_colunas_nomes = {coluna: coluna[:4] + ' Exp./L' if not ind%2 else coluna[:4] + ' Vl.Dolar' for ind, coluna in enumerate(colunas_nomes_anos) }
```

```python
df_ExpVinho.rename(dict_colunas_nomes, axis=1)
```

```python
df_ExpVinho = df_ExpVinho.set_index('Id')
```

```python
df_ExpVinho2021 = df_ExpVinho.rename(dict_colunas_nomes, axis=1).iloc[:,[0,-2,-1]]
df_ExpVinho2021
```

```python
df_TopExp_2021 = df_ExpVinho2021.sort_values(by='2021 Vl.Dolar', axis=0, ascending=False)
df_TopExp_2021.reset_index(drop=True, inplace=True)
df_TopExp_2021
```

```python
df_TopExp_2021.head(10)
```

```python
plt.figsize=(20, 5)
sns.barplot(x='País', y='2021 Exp./L', palette='blend:purple,green', data=df_TopExp_2021.head(10))
plt.xticks(rotation=90)
```

```python
df_ExpPerc = df_TopExp_2021['2021 Exp./L'].sum()
```

```python
df_TopExp_2021['Porcentagem'] = (df_TopExp_2021['2021 Exp./L'] / df_ExpPerc) * 100
```

```python
df_TopExp_2021
```

```python
df_TopExp_2021['Porcentagem'] = df_TopExp_2021['Porcentagem'].round(2)
df_TopExp_2021.head(20)
```
